Focando no banco de dados disponível, o arquivo com as notas dadas por 138493 usuários à 27278 filmes pode ser representado por uma matriz: RATING[138493, 27278]. Para fazer um paralelo com o exemplo dado, a matiz RATING seria o equivalente à matriz [57 4](resultado do exemplo sobre o gosto de Luiza). Baseado nesse paralelo, a matriz RATING[138493, 27278] pode ser reescrita como o produto de duas matrizes USER[138493, features] e FILM[Features, 27278]

A variável "features" seria (em uma primeira aproximação) a quantitade total dos diferentes genêros de filmes, no nosso caso há 18. Esse valor pode ser modificado futuramente para melhorar o resultado final.

O próximo passo é a pergunta: dado que USER*FILM=RATING, quais os dados conhecidos em cada matriz e como determinar os elementos não conhecidos? 

Resposta-1: As matrizes USER e FILM são totalmente desconhecidas, já a matriz RATING tem alguns elementos conhecidos.

Resposta-2: Uma maneira de determinar todos os valores das três matrizes é através do método interativo. Este método consiste basicamente em preencher as matrizes USER e FILM com valores aleatórios, em seguida fazer a multtiplicação de ambas (vamos chamar a matriz resultante de Predicted_Rating). Os respectivos valores das matrizes RATING e Predicted_Rating devem ser comparados, inicialmente os valores serão totalmente diferentes. Para termos uma matriz Predicted_Rating com valores próximos aos da materiz RATING, devemos mofidicar as matrizes USER e FILM constantemente até que critérios de igualdade sejam satisfeitos. O método será destacado no código.

O resultado final nos dará informações relevantes sobre a compatibilidade entre os usuários e filmes, assim poderemos estimar as avaliações que estão falantando na tabela inicial. 

Somente neste instante poderemos entregar o melhor resultado: indicação de filmes para usuários baseada no histórico de avaliação do mesmo.

Outros métodos podem ser usados para prever as notas que faltam na tabela inicial, mas esse método (que irei empregar) tem a grande vantagem de que no final iremos extrair todas as caracteriscas das partes envolvidas (filmes e usuários) representadas em duas matrizes (que é uma maniera eficaz de representação do ponto de vista computacional). O segundo método de sugestão, pode ser usado para trabalhar com usuários que nunca fizeram avaliações (ou seja, não conhecemos suas preferências). O usuário busca por um filme especifico, e como conhecemos as caracteriscas dos filmes, podemos sugerir filmes com caracteristicas parecidas. O
